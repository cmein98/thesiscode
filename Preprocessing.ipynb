```python
import pandas as pd
import pyreadstat
file_paths = ["data2018.sav",
              "data2019.sav",
              "data2020.sav",
              "data2021.sav",
              "data2022.sav",
              "data2023.sav"]

all_columns = set()

#Loop through each file
for file_path in file_paths:
    # Read the SPSS file and metadata
    df, _ = pyreadstat.read_sav(file_path)
    # Add the columns of the current DataFrame to the set of all columns
    all_columns.update(df.columns)

all_columns = list(all_columns)

#DataFrame to store the combined data
combined_df = pd.DataFrame()

#Loop through each file again
for file_path in file_paths:
    # Read the SPSS file into a DataFrame
    df, _ = pyreadstat.read_sav(file_path)
    # Reindex the DataFrame to include all columns and fill missing values with NaN
    df = df.reindex(columns=all_columns)
    # Append the DataFrame to the combined DataFrame
    combined_df = pd.concat([combined_df, df])

print(combined_df.head(25))
# Check the number of columns and rows
num_columns = combined_df.shape[1]
print("Number of columns:", num_columns)
num_rows = combined_df.shape[0]
print("Number of rows:", num_rows)
```
